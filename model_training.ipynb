```python
from ultralytics import YOLO
import torch
import os

def train_custom_model():
    # 1. Setup Device
    if torch.cuda.is_available():
        device = '0'
    else:
        device = 'cpu'
    
    print(f"Training on: {device}")

    
    model = YOLO("yolo11n.pt") 

    
    results = model.train(
        data=r"E:\insiso_assignment\DATASET1\data.yaml",
        epochs=100,        
        imgsz=640,
        batch=16,          
        patience=15,       
        augment=True,      
        device=device,
        name="concrete_mix_545",
        exist_ok=True,
        val=True
    )

    
    print("\nStarting evaluation on TEST set...")
    metrics = model.val(split='test', name="concrete_test_eval")
    
    # 5. Results
    print(f"\n--- Final Test Results ---")
    print(f"mAP@50: {metrics.box.map50:.4f}")
    print(f"mAP@50-95: {metrics.box.map:.4f}")

if __name__ == '__main__':
    train_custom_model()
```
